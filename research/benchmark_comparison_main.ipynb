```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV file
df_csv = pd.read_csv("combined updated .csv")

# Reshape the data for easier plotting
df_melted = df_csv.melt(id_vars='class', var_name='Model', value_name='Score')

# Plot 1: Individual metric bar plots for each metric
metrics = df_csv['class'].unique()
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

for i, metric in enumerate(metrics):
    sns.barplot(
        data=df_melted[df_melted['class'] == metric],
        x='Model', y='Score', ax=axes[i], palette='Set2'
    )
    axes[i].set_title(f"{metric} by Model")
    axes[i].set_ylim(0, 1)
    axes[i].set_ylabel("Score")

plt.tight_layout()
plt.show()

# Plot 2: Grouped bar chart comparing all metrics per model
plt.figure(figsize=(10, 6))
sns.barplot(data=df_melted, x='Model', y='Score', hue='class', palette='Set1')
plt.title("All Evaluation Metrics by Model")
plt.ylim(0, 1)
plt.ylabel("Score")
plt.legend(title="Metric")
plt.tight_layout()
plt.show()
```
